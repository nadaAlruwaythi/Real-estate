convert to dummies

```python
result=pd.get_dummies(df, columns=["commercial_or_residential", "property_type",'closed','driver_room','duplex','family_or_single','furnished','maid_room','advertiser_type','pool','paid','rent_period','street_direction','for_rent_or_sale','region_name_en','nearest_city_name_en','district_name_en'])
result.head()
```

Model  

```python
X, y = result.drop('price',axis=1), result['price']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge 

scaler = StandardScaler()
#train_size=80%   test=20
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=10)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
print("Training Data has",X_train.shape)
print('Test Data has',X_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled,y_train)
print('Train:',lr.score(X_train_scaled,y_train))
```

```python
lr = LinearRegression() 
lr.fit(X_train, y_train) 
preds = lr.predict(X_test)
test_score = lr.score(X_test, y_test)
print('Train:',lr.score(X_train_scaled,y_train))

print('Testing R^2 score was:', test_score)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

model = RandomForestRegressor(n_jobs=-1)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

train_score = r2_score(y_train, model.predict(X_train))
test_score = r2_score(y_test, model.predict(X_test))

# report results
print('\nTraining R^2 score was:', train_score)

print('\nTesting R^2 score was:', test_score)
```
